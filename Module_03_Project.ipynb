* **SkinThickness:** Triceps skin fold thickness (mm)

* **BMI:** Body mass index (weight in kg/((height in m$)^2$))
* **DiabetesPedigreeFunction:** Diabetes pedigree function (a function which scores likelihood of diabetes based on family history)
* **Age:** Age (in years)
* **Outcome:** Class variable (0 if non-diabetic, 1 if diabetic)

# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
##########################
##########################

## TASK-6 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################


def plot_KNN_error_rate(xdata,ydata):
  error_rate = []
  test_scores = []
  train_scores = []

  ## [REQUIRED] Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
  X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=101)

  for i in range(1,40):
      ## [REQUIRED] Complete the code in the next three lines
      knn = KNeighborsClassifier(n_neighbors=i)
      knn.fit(X_train, y_train)
      pred_i = knn.predict(X_test)

      error_rate.append(np.mean(pred_i != y_test))
      train_scores.append(knn.score(X_train,y_train))
      test_scores.append(knn.score(X_test,y_test))

  plt.figure(figsize=(12,8))
  plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
          markerfacecolor='red', markersize=10)
  plt.title('Error Rate vs. K Value')
  plt.xlabel('K')
  plt.ylabel('Error Rate')
  print()
  ## score that comes from testing on the same datapoints that were used for training
  max_train_score = max(train_scores)
  train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
  print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
  print()
  ## score that comes from testing on the datapoints that were split in the beginning to be used for testing solely
  max_test_score = max(test_scores)
  test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
  print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))

  return test_scores
```

## Modify `plot knn error rate` function

### Subtask:
Update the function to perform K-Fold cross-validation and return cross-validation scores.


**Reasoning**:
The subtask requires modifying the existing function to incorporate K-Fold cross-validation. This involves using StratifiedKFold, iterating through the splits, training and evaluating the model on each fold, and then averaging the results for each k.



```python
from sklearn.model_selection import StratifiedKFold

def plot_KNN_error_rate_kf(xdata, ydata, n_splits=5):
    error_rate = []
    test_scores = []
    train_scores = []
    avg_test_scores = []
    avg_error_rate = []

    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=101)

    for i in range(1, 40):
        fold_error_rate = []
        fold_test_scores = []
        fold_train_scores = []

        for train_index, test_index in kf.split(xdata, ydata):
            X_train, X_test = xdata[train_index], xdata[test_index]
            y_train, y_test = ydata[train_index], ydata[test_index]

            knn = KNeighborsClassifier(n_neighbors=i)
            knn.fit(X_train, y_train)
            pred_i = knn.predict(X_test)

            fold_error_rate.append(np.mean(pred_i != y_test))
            fold_train_scores.append(knn.score(X_train, y_train))
            fold_test_scores.append(knn.score(X_test, y_test))

        avg_error_rate.append(np.mean(fold_error_rate))
        avg_test_scores.append(np.mean(fold_test_scores))
        # Note: Calculating average train score across folds might not be as informative as
        # the average test score in this context, but we can keep track if needed.
        # avg_train_scores.append(np.mean(fold_train_scores))


    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), avg_error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Average Error Rate vs. K Value (K-Fold Cross-Validation)')
    plt.xlabel('K')
    plt.ylabel('Average Error Rate')
    plt.show()

    print()
    # Find max average test score
    max_avg_test_score = max(avg_test_scores)
    max_avg_test_score_ind = [i for i, v in enumerate(avg_test_scores) if v == max_avg_test_score]
    print('Max average test score {} % and k = {}'.format(max_avg_test_score * 100,
                                                          list(map(lambda x: x + 1, max_avg_test_score_ind))))

    return avg_test_scores
```

**Reasoning**:
The modified function `plot_KNN_error_rate_kf` has been defined. Now, the next step is to call this function with the unchanged data and the scaled data (both Standardized and MinMax scaled) to perform the K-Fold cross-validation experiments as required by the main task.



```python
unchanged_test_scores_kf = plot_KNN_error_rate_kf(unchanged_data.values, diabetes_data['Outcome'].values)
scaled_test_scores_kf = plot_KNN_error_rate_kf(scaled_data, diabetes_data['Outcome'].values)
scaled_test_scores_minmax_kf = plot_KNN_error_rate_kf(scaled_data_minmax, diabetes_data['Outcome'].values)
```

**Reasoning**:
All the classification experiments using K-Fold cross-validation have been performed on the different scaled versions of the data. The next step according to the main task is to present an analysis of the results obtained from these experiments. A good way to analyze the results is to compare the average test scores across different K values for each scaling method by plotting them on the same graph.



```python
plt.figure(figsize=(20, 8))
plt.title('Average Test Accuracy vs. K Value (K-Fold Cross-Validation)')
sns.lineplot(x=range(1, 40), y=unchanged_test_scores_kf, marker='o', label='Unscaled data')
sns.lineplot(x=range(1, 40), y=scaled_test_scores_kf, marker='o', label='Standard Scaled data')
sns.lineplot(x=range(1, 40), y=scaled_test_scores_minmax_kf, marker='o', label='MinMax Scaled data')
plt.xlabel('K')
plt.ylabel('Average Test Accuracy')
plt.legend()
plt.show()
```
